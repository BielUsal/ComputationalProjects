```julia
using Distributions
using Plots
using StatsBase
```

```julia
f(x) = pdf(Cauchy(0,1),x) 
plot(f)
```

```julia
function mh_sampler!(f,T,samples)
    x₀ = rand()
    for i = 1:T
        xₙ= rand(Normal(x₀,2)) 
        if min(1,f(xₙ)/f(x₀)) > rand()
            x₀ = xₙ
        end
    end
    push!(samples,x₀)
end
```

```julia
samples = Float64[]
for i = 1:1000
    mh_sampler!(f,1000,samples)
end
samples=Float64[]
for i = 1:10000
    mh_sampler!(f,1000,samples)
end
```
